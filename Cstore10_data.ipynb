```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# '연령대' 값을 숫자로 치환합니다.
Cstore_rich['연령대'] = Cstore_rich['연령대'].replace({'10대 이하': '10', '20대 이하': '20', '30대 이하': '30', '40대 이하': '40'})

# '성별'과 '연령대'를 결합한 그룹 식별자 생성
Cstore_rich['그룹'] = Cstore_rich['연령대'] + '-' + Cstore_rich['성별']

# '그룹'을 기준으로 '지불 금액'의 평균을 계산합니다.
grouped = Cstore_rich.groupby('그룹')['지불 금액'].mean()

# 그래프를 그리기 위한 데이터 준비
labels = grouped.index
values = grouped.values

# '연령대' 값을 기준으로 그룹 내 '성별' 값을 정렬합니다.
labels = sorted(labels, key=lambda x: (int(x.split('-')[0]), {'남': 0, '여': 1, '미상': 2}[x.split('-')[1]]))

# '성별' 값을 기준으로 색상 할당
gender_colors = {'남': 'blue', '여': 'red', '미상': 'purple'}
colors = [gender_colors[gender.split('-')[1]] for gender in labels]

# 그래프 그리기
x = np.arange(len(labels))

fig, ax = plt.subplots()
scatter = ax.scatter(x, values, c=colors, s=100)

# 축, 라벨, 제목 설정
ax.set_xlabel('성별 및 연령대')
ax.set_ylabel('지불 금액 평균')
ax.set_title('성별 및 연령대에 따른 지불 금액 평균')

# x 축 레이블 설정
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right')

# 그래프 출력
plt.tight_layout()  # 그래프 사이 간격 조정
plt.show()
```
